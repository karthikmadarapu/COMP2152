```python
# Dictionary storing workout statistics (dictionary with tuple values)

workout_stats = {
    "Alex" : (30, 45, 20),
    "Jamie": (40, 30, 25),
    "Taylor": (20, 50, 40)
}
```

```python
# Calculating total workout minutes for each friend
for friend, minutes in list(workout_stats.items()):
    total_minutes = sum(minutes)
    workout_stats[f"{friend}_Total"] = total_minutes
```

```python
# 2D list: list of lists containing workout minutes
workout_list = []

for friend in ["Alex", "Jamie", "Taylor"]:
    workout_list.append(list(workout_stats[friend]))
```

```python
print("Yoga and Running minutes:")
for row in workout_list:
    print(row[:2])


```

```python
print("Weightlifting minutes (last two friends):")
for row in workout_list[-2:]:
    print(row[2])
```

```python
for friend in ["Alex", "Jamie", "Taylor"]:
    total = workout_stats[f"{friend}_Total"]
    if total >= 120:
        print(f"Great job staying active, {friend}!")
```

```python
name = input("Enter friend's name: ")

if name in workout_stats:
    yoga, running, weights = workout_stats[name]
    total = workout_stats[f"{name}_Total"]

    print(f"{name}'s Workout Stats:")
    print(f"Yoga: {yoga} minutes")
    print(f"Running: {running} minutes")
    print(f"Weightlifting: {weights} minutes")
    print(f"Total: {total} minutes")
else:
    print(f"Friend {name} not found in the records.")
```

```python
totals = {
    "Alex": workout_stats["Alex_Total"],
    "Jamie": workout_stats["Jamie_Total"],
    "Taylor": workout_stats["Taylor_Total"]
}

highest = max(totals, key=totals.get)
lowest = min(totals, key=totals.get)

print(f"Highest total workout minutes: {highest}")
print(f"Lowest total workout minutes: {lowest}")
```
